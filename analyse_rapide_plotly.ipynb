```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots

pd.set_option('display.max_columns', 200)
pd.set_option('display.max_rows', 100)
```

```python
# Chargement du dataset principal
csv_path = 'dataset_fused.csv'
df = pd.read_csv(csv_path)

print(f'Fichier charge: {csv_path}')
print(f'Dimensions: {df.shape[0]} lignes, {df.shape[1]} colonnes')
```

```python
# Apercu rapide
display(df.head())

resume_types = pd.DataFrame({
    'colonne': df.columns,
    'type': df.dtypes.astype(str).values,
    'nb_valeurs_uniques': [df[c].nunique(dropna=True) for c in df.columns]
}).sort_values(by='type')

display(resume_types.head(30))
```

```python
# Statistiques descriptives des colonnes numeriques
num_cols = df.select_dtypes(include=[np.number]).columns.tolist()
cat_cols = df.select_dtypes(exclude=[np.number]).columns.tolist()

print(f'Variables numeriques: {len(num_cols)}')
print(f'Variables non numeriques: {len(cat_cols)}')

display(df[num_cols].describe().T.head(20))
```

```python
# Histogrammes interactifs sur un echantillon de variables numeriques
sample_num_cols = num_cols[:12]
if len(sample_num_cols) == 0:
    print('Aucune colonne numerique disponible.')
else:
    ncols = 3
    nrows = int(np.ceil(len(sample_num_cols) / ncols))
    fig = make_subplots(rows=nrows, cols=ncols, subplot_titles=sample_num_cols)

    for i, col in enumerate(sample_num_cols):
        row = i // ncols + 1
        col_pos = i % ncols + 1
        fig.add_trace(
            go.Histogram(x=df[col], name=col, showlegend=False, nbinsx=30),
            row=row,
            col=col_pos
        )

    fig.update_layout(height=300 * nrows, width=1200, title_text='Distributions (histogrammes) - 12 premieres variables numeriques')
    fig.show()
```

```python
# Boxplots pour detecter rapidement les valeurs extremement atypiques
box_cols = num_cols[:8]
if len(box_cols) == 0:
    print('Aucune colonne numerique disponible.')
else:
    df_box = df[box_cols].melt(var_name='variable', value_name='valeur')
    fig_box = px.box(
        df_box,
        x='variable',
        y='valeur',
        points='outliers',
        title='Boxplots - 8 premieres variables numeriques'
    )
    fig_box.update_layout(xaxis_tickangle=-45, height=550)
    fig_box.show()
```

```python
# Correlation entre variables numeriques
corr_cols = num_cols[:30]  # Limite pour garder un graphique lisible
if len(corr_cols) < 2:
    print('Pas assez de colonnes numeriques pour une matrice de correlation.')
else:
    corr = df[corr_cols].corr(numeric_only=True)
    fig_corr = px.imshow(
        corr,
        text_auto=False,
        color_continuous_scale='RdBu_r',
        zmin=-1,
        zmax=1,
        title='Matrice de correlation (30 premieres variables numeriques)'
    )
    fig_corr.update_layout(height=900, width=900)
    fig_corr.show()
```

```python
# Relations entre variables (scatter matrix)
candidate_cols = ['force_x_rms', 'force_y_rms', 'force_z_rms', 'vib_x_rms', 'vib_y_rms', 'vib_z_rms', 'ae_rms']
scatter_cols = [c for c in candidate_cols if c in df.columns][:5]
color_col = 'tool_condition' if 'tool_condition' in df.columns else None

if len(scatter_cols) < 2:
    print('Pas assez de colonnes cibles presentes pour la scatter matrix.')
else:
    fig_scatter = px.scatter_matrix(
        df,
        dimensions=scatter_cols,
        color=color_col,
        title='Scatter Matrix sur variables capteurs selectionnees'
    )
    fig_scatter.update_layout(height=800, width=1000)
    fig_scatter.show()
```
